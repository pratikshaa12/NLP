### Bag Of Words(BOW)

```python
# example of single document
# without stopwords
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
import pandas as pd

 # single document (',' seperates each document)
string= ["This is an example of bag of words!"]

 # this step will convert text into tokens
vect1= CountVectorizer()
vect1.fit_transform(string)
print("bag of words :", vect1.get_feature_names_out())
```

```python
vect1.vocabulary_
```

##### fit and transform and predict if the  word is present or not. this is widely used for document or subject classification.

```python
c_vect= CountVectorizer()
c_vect.fit(string)
```

```python
string2= ['lets understand is of words is']
c_new_vect= c_vect.transform(string2)
print ("Text present at ", c_new_vect.toarray())

# compare with the indexes
print("original indexes", vect1.get_feature_names_out())
```

```python
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# bag of words using stopwords (u can avoid writing extra steps to remove stopwords)
stpwords= stopwords.words('english')
string=["This is an example of bag of words!"]
vect1=CountVectorizer(stop_words=stpwords)
print(vect1)
```

```python
vect1.fit_transform(string)
print("bag of words:", vect1.get_feature_names_out())
print("vocab       :", vect1.vocabulary_)
```

```python
# using function
def text_matrix(message,countvect):
  terms_doc = countvect.fit_transform(message)
  return pd.DataFrame(terms_doc.toarray(), columns=countvect.get_feature_names_out())
```

```python
message = ['We are slowly making progress in Natural Language Processsing',
           "We will get there","But practice is the only mantra for success"]
c_vect =CountVectorizer()
print("Below matrix is the Bag of Words approach")
text_matrix(message, c_vect)
```

## n-grams

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import word_tokenize
string= ["This is an example of gram!"]
vect1 = CountVectorizer(ngram_range=(1,1))
vect1.fit_transform(string)

vect2= CountVectorizer(ngram_range=(2,2))
vect2.fit_transform(string)


vect3 = CountVectorizer(ngram_range=(3,3))
vect3.fit_transform(string)



vect4 = CountVectorizer(ngram_range=(4,4))
vect4.fit_transform(string)
print("1-gram  :", vect1.get_feature_names_out())
print("2-gram  :", vect2.get_feature_names_out())
print("3-gram  :", vect3.get_feature_names_out())
print("4-gram  :", vect4.get_feature_names_out())




```

### Tf-Idf(Term frequency-Inverse document frequency)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
tfid= TfidfVectorizer(smooth_idf=False)
doc = ["This is an example.", "We will see how it works.","IDF can be confusing"]
doc_vector= tfid.fit_transform(doc)
df= pd.DataFrame(doc_vector.todense(), columns=tfid.get_feature_names_out())
df
```

```python
print(doc_vector)
```

```python
# using function
def text_matrix(message,countvect):
  terms_doc = countvect.fit_transform(message)
  return pd.DataFrame(terms_doc.toarray(), columns=countvect.get_feature_names_out())
```

```python
# we will call the function created earlier
feb_message = ["What is that covid covid",
               "covid is nothing",
               "covisd cases are dropping"]

tf=  TfidfVectorizer()
text_matrix(feb_message, tf)
```

```python
# importance of covid increased based on the occcurence and total document
jul_message = ["What is that covid covid",
               "covid is bad"]
text_matrix(jul_message,tf)
```

#### countvectorizer,TF-IDF,n-grams

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
arr=["car was cleaned by jack",
     "Jack was cleaned by car."]
```

```python
# if u want  to take into account just term frequencies:
vectorizer= CountVectorizer(ngram_range=(2,2))
# the ngram range specifies your ngram configuaration.
X = vectorizer.fit_transform(arr)
# testing the ngram generation
print("Feature Names \n", vectorizer.get_feature_names_out())
print('Array \n',X.toarray())
```

```python
# and now testing TFIDF vectorizer
# u can still specify n-grams here
vectorizer= TfidfVectorizer(ngram_range=(2,2))
X = vectorizer.fit_transform(arr)
# testing the TFIDF value + ngrams
print(X.toarray())
```

```python
# testing TFIDF vectorizer without normalization
# u can still specify n-grams here
vectorizer= TfidfVectorizer(ngram_range=(2,2), norm=None)
X = vectorizer.fit_transform(arr)
print(X.toarray())
```
